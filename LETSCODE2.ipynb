```python
#ARQUIVO CSV valores separada por virgula. criacao e exportação de planilha.
#transforma arquivo txt em planilha
#manipulação de arquivos csv
import csv
with open('brasil_covid.csv', 'r', encoding='utf-8') as arquivo_csv:
    leitor = csv.reader(arquivo_csv)
    header = next(leitor) #ignorado o header
    for linha in leitor:
        if float(linha[2]) > 1: #quando eu quiser isolar um resultado e printar so ele
            print(linha)
```

```python
#fazer mesma coisa só com codigo pynthon

with open('brasil_covid.csv', 'r') as csv_file:
    linhas = csv_file.read()
    linhas = linhas.split('\n')
    for linha in linhas:
        linha = linha.split(',')
        print(linha)
```

```python
#como faz a escrita de um arqui csv no pynthon
with open('users.csv', 'w', encoding='utf-8',) as arquivo_users:
    escritor = csv.writer(arquivo_users)
    escritor.writerow(['nome', 'sobrenome', 'email', 'genero']) #adicionar linhas multiplas
    escritor.writerow(['Pietro', 'Ribeiro', 'pietro@email.com', 'masculino'])
```
